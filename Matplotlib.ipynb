## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product_category = np.array(['furniture', 'technology', 'office supplies'])
sales = np.array([4110451.0, 4744557.50, 3787492.52])
```

## Bar chart/graph
plt.bar(x axis column , y axis column)

```python
plt.bar(product_category,sales)
plt.show()
```

## Properties of plt.bar :
1.width<br>
2.color<br>

## Properties of plt.title, plt.xlabel, plt.ylabel :
1.fontdict = {'fontsize':x , 'fontweight':y, 'color':'colorname'}

```python
plt.bar(product_category, sales, width = 0.5, align = 'center', color = 'teal')

# adding and formatting the title
plt.title('sales\n',fontdict={'fontsize' : 20, 'fontweight' : 5, 'color' : 'green'})

# labeling the axes
plt.xlabel("product category",fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})
plt.ylabel('sales',fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})

# modify the ticks - ticks are the values on the x and y axes
print(plt.xticks())

plt.show()
```

## plt.xticks and plt.yticks :
1. if the space is less for all the elements to be viewed separately in horizontal manner and overlapping is there then we can rorate them by
   plt.xticks(rotation = 90) 

```python
plt.yticks([1,3]) # changes the location of the y markings(ticks) and marks to only 1 and 3 apart from 0
plt.show()
```

```python
plt.yticks([3,6]) # changes the location of the y markings(ticks) and marks to only 3 and 6 apart from 0
plt.show()
```

```python
# giving manual labels(names) to the markings/ticks
plt.yticks([1,3],['harry','tony']) # makes 1 as harry and 3 as tony
```

```python
plt.yticks(np.arange(10,45,5)) # apart from list other things can also be passed as a parameter
plt.show()
```

```python
# converting/modifying the labels/markings to display how we want them to be displayed
ticks = np.arange(0,6000,1000) # 0 to 6 thousand with a step of 1 thousand
labels = ['{}K'.format(i/1000) for i in ticks] # i/1000 is division with floating result, i//1000 is division with integer result
plt.yticks(ticks,labels) # replace the ticks with labels

# labels: This is a new list that will store the labels for the ticks.
# ['{}K'.format(i//1000) for i in ticks]: This is a list comprehension, which is a concise way to create a new list based on an existing iterable (in this case, ticks). Here's what it does:
# for i in ticks: This part iterates over each element i in the ticks list.
# '{}K'.format(i//1000): For each value i in the ticks list, this expression formats it by dividing it by 1000 (i//1000) and then embedding it in a string as '{}K'. The format() method is used to insert the value of i//1000 into the string.
# i//1000: This performs integer division, effectively converting a number like 3000 to 3 (for thousands). So, if i is 3000, the result is '3K'.

plt.show()
```

## Scatter Plot

```python
profit = np.array(['10', '3', '4', '73', '238'])
sales = np.array(['100', '43', '34', '734', '23'])

plt.title('sales\n',fontdict={'fontsize' : 20, 'fontweight' : 5, 'color' : 'green'})

# labeling the axes
plt.xlabel("product category",fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})
plt.ylabel('sales',fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})

plt.scatter(profit,sales, color = 'red')
plt.show()
```

## Properties of plt.scatter() :
1. alpha = lies between 0 to 1 - indicates how transparent or opaque the markings should be, 1 being full opaque <br>
2. s = size of the markers <br>
3. c or color = 'name of the color for the markers'
4. label = 'name of the label' - used in legend

```python
plt.scatter(profit,sales,alpha = 0.5, s = 300)
```

```python
product_category = np.array(['furniture', 'technology', 'office supplies','technology'])
sales = np.array([4744557.51, 4744557.50, 4744557.55,4744557.76])
profit = np.array(['32325', '32323', '33225','34332'])

plt.title('profit vs sales')

# labeling the axes
plt.xlabel("profit",fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})
plt.ylabel('sales',fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})

plt.scatter(profit[product_category == 'technology'],sales[product_category == 'technology'], c = 'brown', s = 120, label = 'technology')
plt.scatter(profit[product_category == 'furniture'],sales[product_category == 'furniture'], c = 'pink', s = 120, label = 'furniture')
plt.legend()
plt.show()
```

## Line Chart

```python
months = np.array(['january', 'february', 'march', 'april', 'may', 'june', 'july'])
sales  = np.array([35, 50, 23, 22, 50, 100, 40])

plt.plot(months, sales)

plt.title('months vs sales')

# labeling the axes
plt.xlabel("months",fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})
plt.ylabel('sales',fontdict={'fontsize' : 10, 'fontweight' : 4, 'color' : 'blue'})

plt.xticks(rotation = 90)

plt.show()
```

## box and whisker chart or box plot

# subplots

```python
sales_technology = np.array([1013.14, 8298.48, 875.51, 22320.83, 9251.6])
sales_furniture = np.array([7170.14, 75237.48, 1443.51, 2420.83, 21098.6])
sales_supplies = np.array([10.14, 1347.48, 11343.51, 220.83, 2098.6])
years = np.array([2010,2011,2012,2013,2014])

fig, ax = plt.subplots()
technology, = ax.plot(years, sales_technology)
technology.set_label("technology sales")

furniture, = ax.plot(years, sales_furniture)
furniture.set_label("furniture sales")
furniture.set_dashes([2,2,2,2])

supplies, = ax.plot(years, sales_supplies)
supplies.set_label("supplies sales")
supplies.set_dashes([2,5,10,3])

plt.legend()
plt.show()
```

```python
sales_technology = np.array([1013.14, 8298.48, 875.51, 22320.83, 9251.6])
sales_furniture = np.array([7170.14, 75237.48, 1443.51, 2420.83, 21098.6])
sales_supplies = np.array([10.14, 1347.48, 11343.51, 220.83, 2098.6])
years = np.array([2010,2011,2012,2013,2014])

fig, ax = plt.subplots(ncols = 3, nrows = 2, sharex = True, sharey = True)
technology, = ax[0][0].plot(years, sales_technology)
technology.set_label("technology sales")

furniture, = ax[0][1].plot(years, sales_furniture)
furniture.set_label("furniture sales")
furniture.set_dashes([2,2,2,2])

supplies, = ax[0][2].plot(years, sales_supplies)
supplies.set_label("supplies sales")
supplies.set_dashes([2,5,10,3])

plt.legend()
plt.show()
```
