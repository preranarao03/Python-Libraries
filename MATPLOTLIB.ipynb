```python
# MATPLOTLIB  is used for data visualization
# used to make charts like bar graph, line graph, pie chart etc.
# makes it easier to understand data
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
#used so that the chart can be rendered properly
```

```python
#example - 
x=[1,2,3,4,5,6,7]
y=[50,51,53,43,23,78,8]
# to plot these values:
plt.plot(x,y)

# we can also set certain attributes such as color, line width etc.
plt.plot(x,y,color='green',linewidth=5,linestyle='dashed')

#we can set the labels to the x-axis and the y-axis
plt.xlabel('Day')
plt.ylabel('Temp')

#To set the title of the chart
plt.title('Weather')
```

```python
#FORMAT STRINGS IN MATPLOTLIB
x=[1,2,3,4,5,6,7]
y=[50,51,53,43,23,78,8]
plt.plot(x,y,'g+') # g+ indicates that the marker points will be shown as '+' signs
plt.plot(x,y,"g+--") 

#the marker can also be represented in words instead of above fashion
plt.plot(x,y,color="green",marker="+",linestyle="dashed",markersize=20)
#color can also be typed in hexadecimal format
#markersize controls the size of the marker
#alpha lies between 0 and 1, controls the opacity of the line color
plt.plot(x,y,color="green",marker="+",linestyle="dashed",markersize=20,alpha=1)
```

```python
#AXES LABELS, LEGEND, GRID
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

#you can plot multiple lines on the same graph
plt.plot(days,max_t,label="Max") #setting label will tell us what the legend has to say
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

#we need to create a legend
plt.legend(loc="best") #we can all set the location of the legend
                       #loc=best means legend will be present where it is best
    
```

```python
# BAR CHART
# values are represented as vertical bars
```

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [45,67,89,90]

plt.bar(company,revenue) #.bar() used to make bar chart
plt.bar(company,profit) # you can also plot this on the same bar graph
plt.xlabel("Companies")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
```

```python
#HISTOGRAMS
# values will be grouped into certain ranges and then frequency of values will be plotted into graph
# EXAMPLE -
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
#plt.hist(blood_sugar) #by default, 10 ranges(bins) are considered
plt.hist(blood_sugar,bins=3,rwidth=0.95) #bins, you can specify the number of ranges you want
                                          #rwidth (relative width), can specify the width between the bars
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95) #you can specify the ranges using bins=[]
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color="green", histtype="step") #you can specify the ranges using bins=[]
#you can also set the color of graph
#histtype, yo can specify the type of graph
#histtype="step" means only outline of graph will be shown, no fill
```

```python
#you can also plot more than one data set on histogram
blood_sugar_men = [23,45,67,78,98]
blood_sugar_women = [45,32,12,78,90]
plt.hist([blood_sugar_men,blood_sugar_women],bins=3,rwidth=0.95,color=['Green','Orange'],
              label=["Men","Women"], orientation="horizontal")
#orientation we can specify if graph should be vertical or horizontal display
plt.legend()
```

```python
#PIE GRAPH
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food","Phone/Internet bills","Car","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels) #syntax for pie chart

plt.axis("equal") #to make the graph look like a perfect circle

plt.show() #used to hide all the extra info given above the pie graph
```

```python
plt.pie(exp_vals,labels = exp_labels, radius=2) #you can modify the radius of the pie to make bigger/smaller
plt.show()
```

```python
plt.pie(exp_vals,labels = exp_labels, autopct='%0.1f%%') # (autopct = )you can display the percentages on the segments to whatever decimal place you want
plt.show()
```

```python
plt.pie(exp_vals,labels = exp_labels, autopct='%0.1f%%', explode=[0,0.1,0,0,0])
plt.show()
#using explode we can bring and display a segment of the circle slightly outwards from the rest 
```

```python
#SCATTER PLOT
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,6,9]
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.scatter(day,no, color="red")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,6,9]
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
#you can give each plot a different color by spcifying the colors in a list
colors = ["r","g","b","r","g","b","r"]
plt.scatter(day,no, c=colors) # c= colors(colors = name of the list), is the parameter to be given
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,6,9]
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
#you can give each plot point a different color by spcifying the colors in a list
colors = ["r","g","b","r","g","b","r"]

#you can also modify the size of each plot point
sizes = [100,2,3,4,21,45,67]
plt.scatter(day,no, c=colors, s=sizes) # c= colors(colors = name of the list), is the parameter to be given
                                       #similarly, size can also be modified
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,6,9]
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15) 

colors = ["r","g","b","r","g","b","r"]

#you can set the opacity of each marker using alpha(ranges from 0 to 1 )
#you can set the type of marker using marker=" "

plt.scatter(day,no, c=colors, alpha=0.4, marker="+") 
plt.show()
```

```python
#STEM PLOTS

x = [1,2,3,4,5,2,6]
y = [2,3,4,5,6,3,2]

plt.stem(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,2,6]
y = [2,3,4,5,6,3,2]

#you can change the line format using linefmt=":" and marker 
plt.stem(x,y,linefmt=":", markerfmt="r+")
plt.show()
```

```python
x = [1,2,3,4,5,2,6]
y = [2,3,4,5,6,3,2]

#you can change where you want the bottom line to be positioned
#ex - you want the bottom line to be positioned at height=9
plt.stem(x,y,linefmt=":", markerfmt="r+",bottom=9)

#you can also change the color of the base bottom line using basefmt
plt.stem(x,y,linefmt=":", markerfmt="r+",bottom=9, basefmt="green",label="python")

#orientation can be changed using orientation
plt.stem(x,y,linefmt=":", markerfmt="r+",bottom=9, basefmt="green",label="python"
            ,orientation="horizontal")
plt.legend()
plt.show()
```

```python
#BOX PLOT AND WHISKER PLOT
# the yellow line in box plot represents the median
# Q1 = First Quartile
# Q2 = Second Quartile = median
# Q3 = Third Quartile
# Q1 to Q3 represents the box plot
# The upper and lower lines part of the box plot is know as whisker 
# Whsiker part represents the min. and max.
# The circles above and below represent outliers
```

```python
x = [10,20,30,45,67,89]

#to change the orientation and make it horizontal or vertical use vert
plt.boxplot(x, vert=False)
plt.show()
```

```python
x = [10,20,30,45,67,89]
plt.boxplot(x)

#we can change the shape of the box plot into V shape using notch
plt.boxplot(x,notch=True)
plt.show()
```

```python
#STACK PLOT
#will display the area, a.k.a area plot
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

#labels
l = ["area1", "area2", "area3"]
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","g","b"])
plt.legend()
plt.show()
```

```python
#you can also add gridlines to the graph
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

#labels
l = ["area1", "area2", "area3"]
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#to add grid
plt.grid()

plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","g","b"])
plt.legend()
plt.show()
```

```python
#STEP PLOT
# graph will be shown in step format

x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

#color can be set
#marker shape can be set
#marker size can be set as ms=
#marker face color can be set as mfc=

plt.step(x,y,color="r",marker="o",ms=10,mfc="g")
plt.show()
```

```python
#SUBPLOTS
# can display multiple graphs in a proper way

x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y,color="r")

plt.pie([1],colors="r")

x1 = [10,20,30,40]
plt.pie(x)

x2 = ["a","s","d","f"]
y2 = [10,20,30,40]
plt.bar(x2,y2)

plt.show()

#the result is that all the graphs are merged 
#therefore we should use subplot to display all graphs properly
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.subplot(2,2,1) #the coordinates are specified as to where we want to locate this graph
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="r")

x1 = [10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)

x2 = ["a","s","d","f"]
y2 = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show()
```
